```python
import pandas as pd

# Cargar el dataset combinado
df = pd.read_csv('C:/Users/natal/OneDrive/Escritorio/Canadian Pr/datasets/combined_data_no_dates.csv')

# Verifica duplicados basados en las columnas clave
duplicados = df.duplicated(subset=['year', 'month', 'owner_org'], keep=False)

# Muestra las filas duplicadas si existen
if duplicados.any():
    print("Filas duplicadas encontradas basadas en las columnas clave:")
    print(df[duplicados])
else:
    print("No se encontraron filas duplicadas basadas en las columnas clave.")

# 1. Verificación de Filas con '_merge = both'
# Muestra algunas filas donde _merge == 'both'
print("Muestras de filas con _merge = 'both':")
print(df[df['_merge'] == 'both'].head(10))  # Muestra las primeras 10 filas donde hubo coincidencia en ambos datasets

# 2. Verificación de Columnas Clave
# Lista de columnas que se utilizaron para hacer el merge (sin incluir 'date')
key_columns = ['year', 'month', 'owner_org']  # Ajusta esta lista según las columnas clave usadas para el merge

# Verificar coincidencias en las columnas clave para las filas donde _merge = 'both'
print("\nVerificación de coincidencias en columnas clave:")
for col in key_columns:
    coincidencias = df[df['_merge'] == 'both'][col + '_x'] == df[df['_merge'] == 'both'][col + '_y']
    print(f"{col}: {coincidencias.mean() * 100:.2f}% de coincidencia")

# 3. Análisis Adicional
# Comparar algunas columnas relevantes entre los datasets
print("\nComparación de algunas columnas clave entre los datasets combinados (donde _merge = 'both'):")
columnas_comparacion = ['reference_number', 'owner_org_title', 'record_created', 'record_modified']  # Ajusta según las columnas importantes

# Muestra las diferencias entre las columnas _x y _y para las columnas seleccionadas
for col in columnas_comparacion:
    diferencias = df[df['_merge'] == 'both'][f'{col}_x'] != df[df['_merge'] == 'both'][f'{col}_y']
    print(f"Diferencias en {col}: {diferencias.sum()} filas no coinciden.")

# Adicionalmente, puedes mostrar algunas filas con diferencias
print("\nEjemplos de filas con diferencias en las columnas clave:")
for col in columnas_comparacion:
    print(f"\nDiferencias en {col}:")
    print(df[(df['_merge'] == 'both') & (df[f'{col}_x'] != df[f'{col}_y'])].head(5))  # Muestra las primeras 5 diferencias
```

```python
import pandas as pd

# Cargar el dataset combinado
df = pd.read_csv('C:/Users/natal/OneDrive/Escritorio/Canadian Pr/datasets/combined_data_no_dates.csv')

# Paso 1: Comprueba la cantidad de valores nulos en cada columna para ver si hay problemas evidentes
print("Valores nulos por columna:")
print(df.isnull().sum())

# Paso 2: Verifica si hay columnas clave (como 'owner_org') que deberían tener valores completos tras el merge
columnas_clave = ['year', 'month', 'owner_org']  # Modifica con las columnas que consideres clave
nulos_columnas_clave = df[columnas_clave].isnull().sum()

print("\nValores nulos en columnas clave:")
print(nulos_columnas_clave)

# Paso 3: Comprueba la cantidad de filas que se encuentran duplicadas, para asegurarte de que no hay duplicados inesperados
duplicados = df.duplicated(subset=columnas_clave, keep=False)
print(f"\nNúmero de filas duplicadas basadas en {columnas_clave}: {duplicados.sum()}")

# Paso 4: Verifica el contenido de algunas filas de muestra que deberían ser combinadas correctamente
muestra_combinada = df.sample(10)  # Muestra aleatoria de 10 filas para verificación visual
print("\nMuestra aleatoria de filas combinadas:")
print(muestra_combinada)

# Paso 5: Si se encuentra la columna "_merge", analiza cuántas filas se combinaron correctamente
if '_merge' in df.columns:
    print("\nRevisión de la columna _merge:")
    print(df['_merge'].value_counts())
else:
    print("\nNo se encontró la columna '_merge' para verificar la combinación exacta.")
```

```python

# Cargar el DataFrame combinado
df_combined = pd.read_csv('C:/Users/natal/OneDrive/Escritorio/Canadian Pr/datasets/combined_data_no_dates.csv')
# Buscar duplicados basados solo en columnas clave ('owner_org')
duplicados = df_combined[df_combined.duplicated(subset=['owner_org'], keep=False)]

# Verificar si hay duplicados en cada categoría de '_merge'
duplicados_both = duplicados[duplicados['_merge'] == 'both']
duplicados_left_only = duplicados[duplicados['_merge'] == 'left_only']
duplicados_right_only = duplicados[duplicados['_merge'] == 'right_only']

# Mostrar resultados
print(f"Total de duplicados encontrados: {len(duplicados)}")
print(f"Duplicados en 'both': {len(duplicados_both)}")
print(f"Duplicados en 'left_only': {len(duplicados_left_only)}")
print(f"Duplicados en 'right_only': {len(duplicados_right_only)}")

# Verificar si hay coincidencias de 'owner_org' en ambas categorías
# ('both' y 'left_only'/'right_only')
coincidencias = duplicados_both.merge(
    df_combined[df_combined['_merge'] != 'both'],
    on=['owner_org'],
    how='inner'
)

print(f"Filas coincidentes en 'both' y 'left_only'/'right_only': {len(coincidencias)}")
```

```python
import pandas as pd

# Cargar el dataset de contratos
df_contracts = pd.read_csv('C:/Users/natal/OneDrive/Escritorio/Canadian Pr/datasets/proactive_publication_-_contracts_-_contracts.csv')

# Asegurarse de que la columna 'contract_date' esté en formato datetime
df_contracts['contract_date'] = pd.to_datetime(df_contracts['contract_date'], errors='coerce')

# Extraer 'year' y 'month' de 'contract_date'
df_contracts['year'] = df_contracts['contract_date'].dt.year
df_contracts['month'] = df_contracts['contract_date'].dt.month

# Revisar si hay valores nulos en 'year' o 'month' y corregir si es necesario
nulos_year = df_contracts['year'].isnull().sum()
nulos_month = df_contracts['month'].isnull().sum()
print(f"Valores nulos en 'year': {nulos_year}")
print(f"Valores nulos en 'month': {nulos_month}")

# Eliminar filas con fechas inválidas si es necesario
df_contracts = df_contracts.dropna(subset=['year', 'month'])

# Verificar la extracción correcta de las fechas
print(df_contracts[['contract_date', 'year', 'month']].head())

# Guardar el dataset corregido en un nuevo archivo CSV
df_contracts.to_csv('C:/Users/natal/OneDrive/Escritorio/Canadian Pr/datasets/contracts_with_dates_corrected.csv', index=False)

print("El dataset de contratos con fechas corregidas ha sido guardado correctamente.")
```
